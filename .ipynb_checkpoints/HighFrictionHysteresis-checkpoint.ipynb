```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
import seaborn as sns
import math
from matplotlib import ticker
```

# Plot New Transient Data: Experiment 1: Higher friction

```python
Data2 = {'Time': [], 'SeaLevel': [], 'GroundedArea': [], 'Area':[], 'MaxHeights': [], 'Velocity': []}
Data3 = {'SeaLevel': [], 'GroundedArea': [], 'Area':[], 'Velocity': []}

for i in range(80, 432):
    dataSeaLevel2     = pd.read_csv('HHFHysteresisData/SeaLevel_' + str(i) + '.csv')
    dataArea2         = pd.read_csv('HHFHysteresisData/Area_' + str(i) + '.csv')
    dataGroundedArea2 = pd.read_csv('HHFHysteresisData/GroundedArea_' + str(i) + '.csv')
    dataVx2           = pd.read_csv('HHFHysteresisData/Vx_' + str(i) + '.csv')
    Data2['Time'].append(i*25 + 2000)
    Data2['SeaLevel'].append(dataSeaLevel2['sea level'][10])
    Vel2 = dataVx2['velocity:0'].tolist()
    Data2['Velocity'].append(Vel2[0])
    Data2['GroundedArea'].append(dataGroundedArea2['Area'][0])
    Data2['Area'].append(dataArea2['Area'][0])
    zs = dataSeaLevel2['Points:2'].tolist()
    MaxHeight = 0.0
    for i in zs:
        if i > MaxHeight:
            MaxHeight = i
    Data2['MaxHeights'].append(MaxHeight)
    
for i in range(0, 412):
    dataSeaLevel3     = pd.read_csv('HHFReversalHysteresisData/SeaLevel_' + str(i) + '.csv')
    dataArea3         = pd.read_csv('HHFReversalHysteresisData/Area_' + str(i) + '.csv')
    dataGroundedArea3 = pd.read_csv('HHFReversalHysteresisData/GroundedArea_' + str(i) + '.csv')
    dataVx3           = pd.read_csv('HHFReversalHysteresisData/Vx_' + str(i) + '.csv')
    #Data3['Time'].append(i*25 + 2000)
    Data3['SeaLevel'].append(dataSeaLevel3['sea level'][10])
    Vel3 = dataVx3['velocity:0'].tolist()
    Data3['Velocity'].append(Vel3[0])
    Data3['GroundedArea'].append(dataGroundedArea3['Area'][0])
    Data3['Area'].append(dataArea3['Area'][0])
    #zs = dataSeaLevel3['Points:2'].tolist()
    #MaxHeight = 0.0
    #for i in zs:
    #    if i > MaxHeight:
    #        MaxHeight = i
    #Data3['MaxHeights'].append(MaxHeight)

df2 = pd.DataFrame(data=Data2)
df3 = pd.DataFrame(data=Data3)

df2['MaxHeights'] = df2['MaxHeights']+80.0
df2['AverageVelocity'] = df2['Velocity']/df2['Area']
df2['Thickness'] = df2['Area']/60000.0
df2['GroundedArea'] = df2['GroundedArea']/10e6

#df3['MaxHeights'] = df3['MaxHeights']+80.0
df3['AverageVelocity'] = df3['Velocity']/df3['Area']
df3['Thickness'] = df3['Area']/60000.0
df3['GroundedArea'] = df3['GroundedArea']/10e6
```

```python
sns.set_theme(style="ticks")
fig, ax = plt.subplots(2, 1, figsize=(10,14), sharex='col', sharey='row')

# Dark Rise: "#ac0e28" , Rumple: "#27408B"

# Plot (0,0)

ax[0].plot(df2.SeaLevel[0:346], df2.GroundedArea[0:346], color="#ac0e28", label="Ice rise", linewidth = 2.0, zorder=0)
ax[0].plot(df2.SeaLevel[345:364], df2.GroundedArea[345:364], color="#27408B", label="Ice rumple", linewidth = 2.0, zorder=0)
ax[0].scatter(df2.SeaLevel[0], df2.GroundedArea[0], color="black", marker='D', s=60, label='Initial state')

ax[0].plot(df3.SeaLevel[0:397], df3.GroundedArea[0:397], color="#ac0e28", linewidth = 2.0, zorder=0)
ax[0].scatter(df3.SeaLevel[0], df3.GroundedArea[0], color="black", marker='s', s=60, label='Reversal')

# Plot (1,0)

ax[1].plot(df2.SeaLevel[0:346], df2.AverageVelocity[0:346], color="#ac0e28", linewidth = 2.0, zorder=0)
ax[1].plot(df2.SeaLevel[345:364], df2.AverageVelocity[345:364], color="#27408B", linewidth = 2.0, zorder=0)
ax[1].scatter(df2.SeaLevel[0], df2.AverageVelocity[0], color="black", marker='D', s=60)

ax[1].plot(df3.SeaLevel[0:397], df3.AverageVelocity[0:397], color="#ac0e28", linewidth = 2.0, zorder=0)
ax[1].scatter(df3.SeaLevel[0], df3.AverageVelocity[0], color="black", marker='s', s=60, label='Reversal')

################################################################################################################

ax[0].plot(df2.SeaLevel[345], df2.GroundedArea[345], marker="o", color="black", markersize=10, label="Transition point")
ax[1].plot(df2.SeaLevel[345], df2.AverageVelocity[345], marker="o", color="black", markersize=10)

ax[1].set_xlabel("Sea level displacement [$m$]", fontsize=30)

ax[0].tick_params(axis="both", labelsize=20)
ax[1].tick_params(axis="both", labelsize=20)

ax[0].set_xlim(-60,170)
ax[1].set_xlim(-60,170)

ax[0].set_yticks(np.arange(0, 25, 4))
ax[0].set_ylim(-1,25)

ax[1].set_yticks(np.arange(200, 440, 40))
ax[1].set_ylim(200,400)

ax[0].set_ylabel("Grounded area [km$\mathregular{^{2}}$]", fontsize=30)
ax[1].set_ylabel("$v_x$ [m a$\mathregular{^{-1}}$]", fontsize=30)

ax[0].yaxis.get_offset_text().set_fontsize(24)
ax[0].legend(fontsize=20, title_fontsize='50', scatterpoints=1, loc='lower left')

ax[0].grid(linestyle=':')
ax[1].grid(linestyle=':')

plt.subplots_adjust(wspace=0.08, hspace=0.1)

ax[0].text(0.03, 0.96, '(a)', size=25, horizontalalignment='center', verticalalignment='center', transform=ax[0].transAxes)
ax[1].text(0.03, 0.96, '(b)', size=25, horizontalalignment='center', verticalalignment='center', transform=ax[1].transAxes)
 
fig.savefig('HighHysteresis.jpg', format='jpg', dpi=700, bbox_inches = "tight")
```
